```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
## create data for the bar plot

# data sizes
m = 30 # rows (population)
n =  6 # columns (features)

# generate data (initialize with zeros)
data = np.zeros((m,n))
```

```python
# here we are jut generating a random data by adding a growing squared number to each column and change the variance
for i in range(n):
    data[:,i] = 30*np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
data
```

```python
# box plot over the 6 columns as the features
plt.boxplot(data)
plt.show()
```

```python
# another way to visualize the box plot by seaborn
sns.boxplot(data=data)
```

```python
# box plot in pandas dataframe
Data_Frame = pd.DataFrame(data, columns=['Feature1','Feature2','Feature3','Feature4','Feature5','Feature6'])
sns.boxplot(data=Data_Frame,orient='h')
plt.show()
```

```python
sns.boxplot(data=Data_Frame,orient='v')
plt.show()
```
